```python
from sklearn.preprocessing import StandardScaler
feature_to_scale = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'texture_se', 'perimeter_se', 'area_se', 'radius_worst', 'texture_worst', 'perimeter_worst', 'area_worst']

scaler = StandardScaler()
df_scaled_array = scaler.fit_transform(df[feature_to_scale])
df_scaled = pd.DataFrame(df_scaled_array, columns=feature_to_scale, index=df.index)

for col in feature_to_scale:
    df[col + '_scaled'] = df_scaled[col]

print(df[[col + '_scaled' for col in feature_to_scale]].head())
```

```python
df.head()
```

```python
df= df.drop(columns=feature_to_scale)
df.head()
```

## Logarithmization

```python
from sklearn.preprocessing import PowerTransformer
features_to_transform = ['radius_mean_scaled', 'texture_mean_scaled', 'perimeter_mean_scaled', 'area_mean_scaled', 'texture_se_scaled', 'perimeter_se_scaled', 'area_se_scaled', 'radius_worst_scaled', 'texture_worst_scaled', 'perimeter_worst_scaled', 'area_worst_scaled']
pt = PowerTransformer(method='yeo-johnson')
df_transformed_array = pt.fit_transform(df[features_to_transform])

df_transformed = pd.DataFrame(df_transformed_array, columns=features_to_transform, index=df.index)
for col in features_to_transform:
    df[col] = df_transformed[col]
```

```python
df.head()
```

```python
def plot_displots_for_features(df, features):
    for feature in features:
        plt.figure(figsize=(10, 5))
        sns.histplot(df, x=feature, bins=30, kde=True)
        plt.title(f'Distribution of {feature}')
        plt.xlabel(feature)
        plt.ylabel('Frequency')
        plt.grid()
        plt.show()
```

```python
plot_displots_for_features(df, features_to_transform)
```

## Model

```python
from sklearn.model_selection import cross_val_predict
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
```

```python
X = df.drop(columns=['diagnosis_codes'])
y = df['diagnosis_codes']
pipeline = Pipeline([
    ('classifier', RandomForestClassifier(n_estimators=100, random_state=42))
])
y_pred = cross_val_predict(pipeline, X, y, cv=5)
```

```python
print(classification_report(y, y_pred))
```
